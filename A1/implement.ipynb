```python
#Including numba to make the program more effective
import numba as nb
from numba import jit
from numba.core.errors import NumbaDeprecationWarning, NumbaPendingDeprecationWarning
import warnings
```

```python
@jit(nopython=True)
def implement(N,v0,mass,task4=0):
    '''
    Parameters:
        N = Number of particles
        v0 = Initial velocity
    
    Output:
        particles = [[posx1, posy1, velx1, vely1, mass1, radius1]
                     [posx2, posy2, velx2, vely2, mass2, radius2]
                     ...
                     [posxN, posyN, velxN, velyN, massN, radiusN]]
        collision_count = Number of collisions for all particles

    Places particles in the grid at random positions.
    '''
    particles = np.zeros((N,6))        #pos x, pos y, vel x, vel y, masse, radius

    if task4 == 0:
        radius = 1/(2*N) 

        x_pos = np.arange(0+radius,1-radius,radius)
        y_pos = np.arange(0+radius,1-radius,radius)

        #Randomize positions
        random_pos_x = np.copy(x_pos)
        np.random.shuffle(random_pos_x)
        random_pos_y = np.copy(y_pos)
        np.random.shuffle(random_pos_y)

        #Random angles:
        if N < 1000:
            theta = np.linspace(0, 2*np.pi, 10000)
        else:
            theta = np.linspace(0, 2*np.pi, N)
        theta_random = np.copy(theta)
        np.random.shuffle(theta_random)

        for i in range(N):
            particles[i, 0] = float(random_pos_x[i])
            particles[i, 1] = float(random_pos_y[i])
            particles[i, 2] = float(v0*np.cos(theta_random[i]))
            particles[i, 3] = float(v0*np.sin(theta_random[i]))
            particles[i, 5] = radius
        
            if len(mass) == 2:
                if i >= N/2:
                    particles[i, 4] = mass[1]
                else:
                    particles[i, 4] = mass[0]
            else:
                particles[i, 4] = mass[0]
    
    else: 
        radius = 1/(2*N)
        x_pos = np.random.uniform(radius+radius/100, 1-radius-radius/100, size=N)
        y_pos = np.random.uniform(radius+radius/100, 0.5, size=N)

        for i in range(N):
            if i == 0:
                particles[i,0] = float(0.5)
                particles[i,1] = float(0.75) 
                particles[i,2] = 0
                particles[i,3] = -5
                particles[i,4] = mass[0]*25 
                particles[i,5] = radius*task4
            else:
                particles[i, 0] = float(x_pos[i])
                particles[i, 1] = float(y_pos[i])
                particles[i, 2] = 0
                particles[i, 3] = 0
                particles[i, 4] = mass[0]
                particles[i, 5] = radius


    collision_count = np.zeros(N)

    return particles, collision_count
```
