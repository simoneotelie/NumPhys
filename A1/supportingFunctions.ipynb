```python
def updateParticles(particles, particles_involved_index, timestep):
    '''
    Parameteres:
        particles = [[posx,posy,velx,vely,mass,radius]
                  [...]]
        particles_involved_index = The indeces of the particles involved in collision
        timestep = The last calculated timestep
    
    Output:
        particles = [[posx_updates,posy_updates,velx,vely,mass,radius]
                     [...]]
    
    This function updates all the particles that don't collide. 
    '''
    
    if len(particles_involved_index) == 2:
        for i in range(len(particles)):
            if i != particles_involved_index[0] and i != particles_involved_index[1]:
            
                posx = particles[i,0]
                posy = particles[i,1]
                velx = particles[i,2]
                vely = particles[i,3]

                radius = particles[i,5]

                posx_new = posx + velx*timestep
                posy_new = posy + vely*timestep

                if posx_new < radius or posx_new > 1 - radius or posy_new < radius or posy_new > 1 - radius: #Then it is a collision waiting to happen
                    posx_new = posx 
                    posy_new = posy
                

                particles[i,0] = posx_new
                particles[i,1] = posy_new
        
    elif len(particles_involved_index) == 1:
        for i in range(len(particles)):
            if i != particles_involved_index[0]:
                
                posx = particles[i,0]
                posy = particles[i,1]
                velx = particles[i,2]
                vely = particles[i,3]

                radius = particles[i,5]

                posx_new = posx + velx*timestep
                posy_new = posy + vely*timestep


                if posx_new < radius or posx_new > 1 - radius or posy_new < radius or posy_new > 1 - radius:
                    posx_new = posx 
                    posy_new = posy

                particles[i,0] = posx_new
                particles[i,1] = posy_new
    
    return particles
```
